```python
import pandas as pd

df = pd.read_csv("kc_house_data.csv", sep=',')
df.head()
```

```python
import matplotlib.pyplot as plt

plt.hist(df['price'], bins=50)
plt.title('Распределение цены')
plt.xlabel('Стоимость')
plt.ylabel('Количество домов');
-- Дома пропорциональны стоимости
```

```python
plt.boxplot(df['price'])
plt.xlabel('Цена')
plt.ylabel('Количество домов')
plt.title('Стоимость недвижимости')
-- Чем ниже чена, тем больше домов по данной цене
```

```python
plt.hist(df['sqft_living'], bins=50)
plt.xlabel('Квадратура жилой площади')
plt.ylabel('Количество домов')
plt.title('Распределение квадратуры жилой')
-- Наибольшее количество домов в диапазоне "пиков" графика
```

```python
plt.hist(df['yr_built'], bins=50)
plt.xlabel('Год постройки')
plt.ylabel('Количество домов')
plt.title('Распределение года постройки')
-- нет равномерной тенденции строительства домов, наименьшее количество построенных домов в конце 1930х, пики строительства в конце 1970х и 2010х
```

```python
waterfront_counts = df['waterfront'].value_counts()
plt.figure(figsize=(2, 2))
plt.pie(waterfront_counts, labels=['Нет', 'Да'], autopct='%1.1f%%', startangle=90)
plt.title('Распределение домов от наличия вида на набережную')
-- вид на набережную только у 0,8% домов
```

```python
df['floors'].value_counts().sort_index().plot(kind='bar')
plt.xlabel('Количество этажей')
plt.ylabel('Количество домов')
plt.title('Распределение этажей домов')
-- Преобдадают одноэтажные постройки
```

```python
import seaborn as sns
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='condition', hue='floors')
plt.xlabel('Состояние дома')
plt.ylabel('Количество домов')
plt.title('Многорядная столбчатая диаграмма для состояния и этажей')
plt.legend(title='Количество этажей')
-- Больнинство домов находятся в среднем состоянии (касается всех этажей), также в самом лучшем состоянии находятся одноэтажные дома, однако домов, в которых 2,5 этажа нет в аварийном и плохом состоянии.
```

```python
plt.figure(figsize=(2, 3))
sns.countplot(data=df, x='condition')
plt.xlabel('Состояние дома')
plt.ylabel('Количество домов')
plt.title('Распределение состояния домов')
-- Наибольшее количество домов в среднем и хорошем состоянии
```
